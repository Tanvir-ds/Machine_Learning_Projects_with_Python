```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#importing datasets
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
#finding missing values
df.isnull().sum()
```

```python
#finding duplicates values
df.duplicated().sum()
```

```python
#Visualization(x-axis is always Independent and y-axis is always Dependent Variables)
plt.xlabel("Year")
plt.ylabel("Per Capita Income(US$)")
plt.scatter(df["year"],df["per capita income (US$)"])
plt.show()
```

```python
#Extarcting Depent and Independent variables
x = df[["year"]]
y = df["per capita income (US$)"]
```

```python
x
```

```python
y
```

```python
#splitting the datasets into training and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
#Training the model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
#Seeing the test accuracy
lin_reg.score(x_test,y_test)
```

```python
#predicting the test set result
y_predicted = lin_reg.predict(x_test)
y_predicted
```

```python
y_test
```

```python
#plotting the best fit line
plt.xlabel("Year")
plt.ylabel("Per Capita Income(US$)")
plt.scatter(df["year"],df["per capita income (US$)"])
plt.plot(x_test,y_predicted,color="green")
plt.show()
```
